```python
import pandas as pd
import seaborn as sns

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "Times New Roman",  # Set font to Times New Roman
    # "font.size": 10,  # Adjust font size as needed
    # "axes.labelsize": 12,  # Axis labels
    "xtick.labelsize": 12,  # X-axis tick labels
    "ytick.labelsize": 12,  # Y-axis tick labels
    "legend.fontsize": 10   # Legend text
})

# open csv file and load into pandas dataframe
GRID_FILE = "summary//grid4x4//all.csv"
INGO_FILE = "summary//ingo//all.csv"


grid_df = pd.read_csv(GRID_FILE)
ingo_df = pd.read_csv(INGO_FILE)
```

```python
# rename the columns
grid_df = grid_df.rename(columns={'model_name': 'Models'})
ingo_df = ingo_df.rename(columns={'model_name': 'Models'})
```

```python
grid_case_mapping = {
    "robust_morning_1": 1,
    "robust_morning_2": 2,
    "robust_morning_3": 3,
    "robust_morning_4": 4,
    "robust_morning_5": 5,
    "robust_evening_1": 6,
    "robust_pse_1": 7,
    "robust_sensor": 8,
}

grid_df["case"] = grid_df["case"].map(grid_case_mapping)
```

```python
# Create a bar plot using seaborn
plt.figure(figsize=(4, 3))
sns.barplot(x='case', y='recovery_time', hue='Models', hue_order=['dqn', 'mplight', 'colight'], data=grid_df)

# Add title and labels
plt.title('Grid 4x4 Recovery Times')
plt.xlabel('Cases')
plt.ylabel('Recovery Time (hour)')
plt.gca().set_facecolor('whitesmoke')
plt.grid(color='gray', axis='y', linestyle='-', linewidth=0.5, alpha=0.3)

# Show the plot
plt.savefig('summary\grid.pdf', bbox_inches='tight')
plt.show()
```

```python
ingo_case_mapping = {
    "ingo_morning": 1,
    "ingo_noon": 2,
    "ingo_evening": 3,
    "ingo_sensor": 4,
}

ingo_df["case"] = ingo_df["case"].map(ingo_case_mapping)
```

```python
# Create a bar plot using seaborn
plt.figure(figsize=(4, 3))
sns.barplot(x='case', y='recovery_time', hue='Models', hue_order=['dqn', 'colight'], data=ingo_df)

# Add title and labels
plt.title('Ingolstadt Recovery Times')
plt.xlabel('Cases')
plt.ylabel('Recovery Time (hour)')
plt.gca().set_facecolor('whitesmoke')
plt.grid(color='gray', axis='y', linestyle='-', linewidth=0.5, alpha=0.3)

# Show the plot
plt.savefig('summary\ingo.pdf', bbox_inches='tight')
plt.show()
```
